```python
import numpy as np
```

```python

print("--- Exercício 1 ---")
ex1 = np.arange(10)  
ex1[5:9] = 0         
print(ex1)

print("\n--- Exercício 2 ---")
ex2 = np.array([[10, 11], [20, 21], [30, 31]])
print("Shape:", ex2.shape)
print("2ª linha:", ex2[1]) 

print("\n--- Exercício 3 ---")
ex3 = np.array([[1, 2], [3, 4], [5, 6]])
print("2ª coluna:", ex3[:, 1]) # O ':' significa 'todas as linhas', o '1' é a 2ª coluna

print("\n--- Exercício 4 ---")
ex4 = np.reshape(np.arange(20), (4, 5))
print("Matriz:\n", ex4)
print("3ª linha:", ex4[2])

print("\n--- Exercício 5 ---")
ex5 = np.reshape(np.arange(20), (4, 5))
print("Recorte (Linhas 1-2, Colunas 2-3):\n", ex5[0:2, 1:3])

print("\n--- Exercício 6 ---")
ex6 = np.reshape(np.arange(20), (4, 5))
# [1:3] pega linhas 1 e 2 | [0:3] pega colunas 0, 1 e 2
print("Recorte (Linhas 2-3, Colunas 1-3):\n", ex6[1:3, 0:3])
```
