```python

inventario_tech = {
    'Componente': ['SSD 1TB M.2', 'Memória RAM 16GB', 'Placa de Vídeo RTX'],
    'Preço_Custo': [350, 220, 1850],
    'Unidades': [12, 30, 4]
}

df_estoque = pd.DataFrame(inventario_tech)

lista_componentes = df_estoque['Componente']
pecas_alto_valor = df_estoque[df_estoque['Preço_Custo'] > 300]

print("--- Catálogo de Componentes ---")
print(lista_componentes)

print("\n--- Itens com Custo Superior a R$ 300 ---")
print(pecas_alto_valor)
```

```python
especificacoes_tecnicas = {
    'Hardware': ['Teclado Mecânico', 'Mouse Gamer', 'Monitor 180Hz'],
    'Valor_Unitario': [250, 120, 1100],
    'Qtd_Disponivel': [15, 40, 8]
}

df_tecnologia = pd.DataFrame(especificacoes_tecnicas)

lista_itens = df_tecnologia['Hardware']

itens_premium = df_tecnologia[df_tecnologia['Valor_Unitario'] > 200]

print("Listagem de Hardware:")
print(lista_itens)

print("\nEquipamentos com valor acima de 200 reais:")
print(itens_premium)
```
